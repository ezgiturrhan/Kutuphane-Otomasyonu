```python


# Gerekli ktphaneler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
import xgboost as xgb
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, roc_curve, auc
import shap
shap.initjs()
import warnings
warnings.filterwarnings('ignore')
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10,6)

# ============================================================
# 1. VER YKLEME
# ============================================================
data = load_breast_cancer()
X = pd.DataFrame(data.data, columns=data.feature_names)
y = pd.Series(data.target, name='target')

print("lk 5 satr (X):")
display(X.head())
print("\nHedef deiken (y) ilk 5 satr:")
display(y.head())
print("\nVeri Boyutu: ", X.shape)
print("Snf Dalm:\n", y.value_counts())

# ============================================================
# 2. VER KALTE KONTROLLER
# ============================================================
# Eksik deer
missing_values = X.isnull().sum()
if missing_values.sum() > 0:
    X.fillna(X.mean(), inplace=True)
    print("Eksik deerler mean ile dolduruldu.")
else:
    print("Eksik deer bulunamad.")

# Aykr deer (IQR)
Q1 = X.quantile(0.25)
Q3 = X.quantile(0.75)
IQR = Q3 - Q1
outliers = ((X < (Q1 - 1.5*IQR)) | (X > (Q3 + 1.5*IQR))).sum()
print("\nAykr deer says stun baznda:")
display(outliers)

plt.figure(figsize=(15,6))
sns.boxplot(data=pd.melt(X), x='variable', y='value')
plt.xticks(rotation=90)
plt.title("Boxplot Analizi (Stun Bazl)")
plt.show()

# Veri tipleri
num_cols = X.select_dtypes(include=np.number).shape[1]
cat_cols = X.select_dtypes(exclude=np.number).shape[1]
print(f"Saysal deiken says: {num_cols}, Kategorik deiken says: {cat_cols}")

# ============================================================
# 3. EDA
# ============================================================
# statistiksel zellikler
eda_stats = X.describe().T
eda_stats['median'] = X.median()
eda_stats['Q1'] = X.quantile(0.25)
eda_stats['Q3'] = X.quantile(0.75)
eda_stats = eda_stats[['mean','median','min','max','std','Q1','Q3']]
print("----- statistiksel zellikler -----")
display(eda_stats)

# Korelasyon Matrisi
plt.figure(figsize=(12,10))
corr_matrix = X.corr()  # Pearson korelasyonu
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5)
plt.title("Korelasyon Matrisi (Pearson)")
plt.show()

# En yksek korelasyonlu 3 ift
upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
top3_corr = upper_tri.unstack().dropna().sort_values(ascending=False).head(3)
print("\nEn yksek korelasyonlu 3 ift zellik:")
display(top3_corr)


# Boxplot tm stunlar
plt.figure(figsize=(15,6))
sns.boxplot(data=pd.melt(X), x='variable', y='value')
plt.xticks(rotation=90)
plt.title("Boxplot Analizi")
plt.show()

# ============================================================
# 4. VER LEKLENDRME
# ============================================================
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print("leklendirme tamamland.")

# ============================================================
# 5. VER SETNN BLNMES
# ============================================================
X_train_full, X_test, y_train_full, y_test = train_test_split(X_scaled, y, test_size=0.2, stratify=y, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_train_full, y_train_full, test_size=0.125, stratify=y_train_full, random_state=42)

print("Train:", X_train.shape, "Validation:", X_val.shape, "Test:", X_test.shape)

# ============================================================
# 6. PCA & LDA
# ============================================================
# PCA
pca_full = PCA()
pca_full.fit(X_train)
cum_var = np.cumsum(pca_full.explained_variance_ratio_)
n_components = max(2, np.argmax(cum_var >= np.mean(pca_full.explained_variance_ratio_)) + 1)
pca = PCA(n_components=n_components)
X_train_pca = pca.fit_transform(X_train)
X_val_pca = pca.transform(X_val)
X_test_pca = pca.transform(X_test)

plt.figure()
plt.plot(cum_var, marker='o')
plt.axhline(y=np.mean(pca_full.explained_variance_ratio_), color='r', linestyle='--')
plt.xlabel("Bileen Says")
plt.ylabel("Kmlatif Varyans")
plt.title("PCA Explained Variance")
plt.show()

plt.figure()
plt.scatter(X_train_pca[:,0], X_train_pca[:,1], c=y_train, cmap='viridis', edgecolor='k')
plt.xlabel("PC1"); plt.ylabel("PC2"); plt.title("PCA 2D Scatter (Train)")
plt.show()

# LDA
lda = LDA(n_components=1)
X_train_lda = lda.fit_transform(X_train, y_train)
X_val_lda = lda.transform(X_val)
X_test_lda = lda.transform(X_test)

plt.figure()
plt.scatter(X_train_lda, np.zeros_like(X_train_lda), c=y_train, cmap='viridis', edgecolor='k')
plt.xlabel("LD1"); plt.title("LDA Scatter (Train) - Binary")
plt.show()

# ============================================================
# 7 & 8. MODEL KURULUMU & VALIDATION
# ============================================================
models = {
    "LogisticRegression": LogisticRegression(max_iter=500),
    "DecisionTree": DecisionTreeClassifier(),
    "RandomForest": RandomForestClassifier(),
    "XGBoost": xgb.XGBClassifier(eval_metric='logloss', use_label_encoder=False),
    "GaussianNB": GaussianNB()
}

datasets = {
    "Ham": (X_train, X_val, X_test),
    "PCA": (X_train_pca, X_val_pca, X_test_pca),
    "LDA": (X_train_lda, X_val_lda, X_test_lda)
}

performance = []
for data_name, (X_tr, X_v, X_te) in datasets.items():
    for model_name, model in models.items():
        model.fit(X_tr, y_train)
        y_pred_val = model.predict(X_v)
        y_proba_val = model.predict_proba(X_v)[:,1] if hasattr(model,"predict_proba") else model.decision_function(X_v)
        acc = accuracy_score(y_val, y_pred_val)
        prec = precision_score(y_val, y_pred_val)
        rec = recall_score(y_val, y_pred_val)
        f1 = f1_score(y_val, y_pred_val)
        roc = roc_auc_score(y_val, y_proba_val)
        performance.append([data_name, model_name, acc, prec, rec, f1, roc])

perf_df = pd.DataFrame(performance, columns=['Dataset','Model','Accuracy','Precision','Recall','F1','ROC-AUC'])
print("===== VALIDATION PERFORMANS TABLOSU =====")
display(perf_df.sort_values(by='Accuracy', ascending=False))

# ============================================================
# 9. TEST PERFORMANSI
# ============================================================
best_row = perf_df.loc[perf_df['Accuracy'].idxmax()]
best_dataset = best_row['Dataset']
best_model_name = best_row['Model']
print(f"\nEn iyi model: {best_model_name} ({best_dataset})")

X_tr_best, X_v_best, X_te_best = datasets[best_dataset]
best_model = models[best_model_name]
best_model.fit(X_tr_best, y_train)
y_test_pred = best_model.predict(X_te_best)
y_test_proba = best_model.predict_proba(X_te_best)[:,1] if hasattr(best_model,"predict_proba") else best_model.decision_function(X_te_best)

# Test Metrikleri
acc = accuracy_score(y_test, y_test_pred)
prec = precision_score(y_test, y_test_pred)
rec = recall_score(y_test, y_test_pred)
f1 = f1_score(y_test, y_test_pred)
roc = roc_auc_score(y_test, y_test_proba)
print(f"Accuracy: {acc:.4f}, Precision: {prec:.4f}, Recall: {rec:.4f}, F1: {f1:.4f}, ROC-AUC: {roc:.4f}")

# Confusion Matrix
cm = confusion_matrix(y_test, y_test_pred)
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix")
plt.show()

# ROC Curve
fpr, tpr, _ = roc_curve(y_test, y_test_proba)
plt.figure(figsize=(6,4))
plt.plot(fpr, tpr, label=f'AUC={roc:.2f}')
plt.plot([0,1],[0,1],'k--')
plt.xlabel("False Positive Rate"); plt.ylabel("True Positive Rate"); plt.title("ROC Curve")
plt.legend()
plt.show()

# ============================================================
# 10. XAI - SHAP
# ============================================================
# Ham veri SHAP
feature_names = X.columns
if best_model_name in ["RandomForest","XGBoost"]:
    explainer = shap.TreeExplainer(best_model)
else:
    explainer = shap.Explainer(best_model, X_tr_best)

shap_values = explainer(X_te_best)
shap.summary_plot(shap_values, features=X_te_best, feature_names=feature_names)
shap.plots.bar(shap_values)

# PCA ve LDA SHAP
def safe_shap_plot(model, X_tr, X_te, feature_names, title):
    try:
        explainer = shap.Explainer(model, X_tr)
        shap_values = explainer(X_te)
        shap.summary_plot(shap_values, features=X_te, feature_names=feature_names, show=False)
        plt.title(title)
        plt.show()
        shap.plots.bar(shap_values)
    except Exception as e:
          print(f"Warning: {title} SHAP hatas olutu -> {e}")

# PCA
model_pca = LogisticRegression(max_iter=500).fit(X_train_pca, y_train)
safe_shap_plot(model_pca, X_train_pca, X_test_pca, [f"PC{i+1}" for i in range(X_test_pca.shape[1])], "PCA SHAP")

# LDA
model_lda = LogisticRegression(max_iter=500).fit(X_train_lda, y_train)
safe_shap_plot(model_lda, X_train_lda, X_test_lda, ["LD1"], "LDA SHAP")
```
